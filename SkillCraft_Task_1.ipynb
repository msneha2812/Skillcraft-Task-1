```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Simulated dataset (you can replace with your CSV file)
data = {
    'square_feet': [1000, 1500, 1800, 2400, 3000, 3500, 4000],
    'bedrooms': [2, 3, 3, 4, 4, 5, 5],
    'bathrooms': [1, 2, 2, 3, 3, 4, 4],
    'price': [150000, 200000, 230000, 310000, 360000, 420000, 480000]
}

df = pd.DataFrame(data)

print("... Sample Data Preview ...")
print(df.head())
print("...")

# Step 2: Features and target
X = df[['square_feet', 'bedrooms', 'bathrooms']]
y = df['price']

# Step 3: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("... Training and Testing Data Sizes ...")
print("Training samples:", len(X_train))
print("Testing samples:", len(X_test))
print("...")

# Step 4: Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Step 5: Predict
y_pred = model.predict(X_test)

# Step 6: Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("... Model Evaluation Metrics ...")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")
print("...")

print("... Model Coefficients and Intercept ...")
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
print("...")

# Step 7: Prediction vs Actual
print("... Predicted vs Actual Prices ...")
for actual, predicted in zip(y_test, y_pred):
    print(f"Actual: ₹{actual:,.0f} | Predicted: ₹{predicted:,.0f}")
print("...")

# Step 8: Plotting
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, color='green', s=100, marker='o', label='Predicted')
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted House Prices")
plt.legend()
plt.grid(True)
plt.show()
```
